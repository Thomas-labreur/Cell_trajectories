```python
from Organize_All_Data_v2 import (
    OUTDATAPATH_IMARIS, OUTDATAPATH_FIJI, UNWANTED, process_data, process_mega
)
import shutil, os , re
import pandas as pd

def load_and_sort_csv_files(folder: str):
    # Regex pour extraire le numéro après "Scene_"
    scene_regex = re.compile(r"Scene[-_](\d+)")

    par_cellule = []
    position_relative = []
    mega= []

    for xp in os.listdir(folder):
        for fname in os.listdir(os.path.join(folder, xp)):
            if not fname.endswith(".csv"):
                continue

            match = scene_regex.search(fname)

            scene_num = int(match.group(1)) if match else 0
            full_path = os.path.join(folder, xp, fname)

            if fname.endswith("Par_Cellule.csv"):
                par_cellule.append((scene_num, full_path))
            elif fname.endswith("Position_Relative.csv"):
                position_relative.append((scene_num, full_path))
            elif 'MEGA' in fname:
                mega.append(full_path)

    # Tri par numéro de scène
    par_cellule.sort(key=lambda x: x[0])
    position_relative.sort(key=lambda x: x[0])
    mega.sort()

    # Retourne seulement les chemins triés
    par_cellule = [pd.read_csv(p) for _, p in par_cellule]
    position_relative = [pd.read_csv(p) for _, p in position_relative]
    mega = [pd.read_csv(p) for p in mega ]

    if mega and 'SEXE' in mega[0].columns:
        mega = [m.drop(columns=['SEXE', 'IMC', 'TRAITEMENT']) for m in mega]

    return par_cellule, position_relative, mega
```

```python
from Organize_All_Data_v2 import DATE_REGEX, CLASS_REGEX, SCENE_REGEX, AGE_REGEX, PATHINFOTABLE, check_issues, isfile, relative_positions, open_imaris_csv, open_metadata
import numpy as np
import pandas as pd
import re, os

def add_metadata_to_dataframe(df: pd.DataFrame, filename: str, df_meta: pd.DataFrame):

    # Ajoute la colonne CLASSE
    res_regex = re.search(CLASS_REGEX, filename)
    class_name = res_regex.group(0) if res_regex else np.nan
    df['CLASSE'] = class_name

    # Ajoute la colonne SCENE
    res_regex = re.search(SCENE_REGEX[1], filename)
    scene = int(res_regex.group(0)) if res_regex else np.nan
    df['SCENE'] = scene

    # Ajout de la date
    res_regex = re.search(DATE_REGEX[0], filename)
    date= res_regex.group(0) if res_regex else np.nan
    df['DATE'] = date
    print(date)

    # Ajout de l'age avec conversion en semaines
    age = df_meta.loc[(df_meta["Date"] == date ) & (df_meta['Scene'] == scene), ['Age']]
    res_regex = re.search(AGE_REGEX, str(age))
    if res_regex:
        age, unit = res_regex.group(1), res_regex.group(2)
        factor =  4.34524 if unit=='mois' else 52.1429 if unit=='an' else 1
        age = round(factor*int(age))

    # Si l'age n'est pas spécifié ou mal spécifié, affiche un avertissement
    else:
        age = np.nan
        print("_________________________________________________________________")
        print("            /!\\ ATTENTION PROBLEME D'AGE' /!\\")
        print(f"Age non spécifié dans le fichier {os.path.basename(PATHINFOTABLE)}")
        print(f"Expérience {date} | Scène {scene} ")
        print("_________________________________________________________________")
    df['AGE'] = age

    # Ajout du sexe, IMC et Traitement
    df['SEXE'] = df_meta.loc[(df_meta["Date"] == date ) & (df_meta['Scene'] == scene), ['Sexe']].values[0][0]
    df['IMC'] = df_meta.loc[(df_meta["Date"] == date ) & (df_meta['Scene'] == scene), ['IMC']].values[0][0]
    df['TRAITEMENT'] = df_meta.loc[(df_meta["Date"] == date ) & (df_meta['Scene'] == scene), ['Traitement']].values[0][0]

    return df

def get_fiji_pos_dataframe(scene_dir: str, xp: str, scene: int) -> pd.DataFrame: 
    """
    Créée un dataframe en concaténant les csv du dossier FIJI fourni, en vérifiant les problèmes
    de doublons et saut dans le temps, et en ajoutant les colonnes de positions relatives.

    Arguments:
        scene_dir (str): chemin vers le dossier de la scène contenant les csv
        xp (str): date de l'expérience
        scene (int): numéro de la scène

    Renvois: 
        pd.DataFrame: le dataframe décri ci-dessus
    """

    list_df, list_fnames = [], []
    for f in os.listdir(scene_dir):
        # Obtenir la liste des csv du répertoire scene_dir
        path = os.path.join(scene_dir, f)
        if isfile(path) and path.endswith(".csv"):

            # Lecture du fichier csv (éventuellement avec le séparateur européen)
            df_csv = pd.read_csv(path, encoding="latin")
            df_csv = pd.read_csv(path, encoding = 'latin', sep = ";") if len(df_csv.columns) != 7 else df_csv
            df_csv.columns = ['TrackID', 'Time', 'PositionX', 'PositionY', 'Distance_µm', 'Speed_µm/min', 'PixelValue']

            # Ajuste les valeurs de TrackID
            if "à" in path:
                diff_trackID = int(re.findall(r'\d+', path)[-1]) - df_csv['TrackID'].max()
                df_csv['TrackID'] += diff_trackID

            # Remet à 0 des premiers temps de chaque cellule (-1 à 0)
            df_csv.loc[(df_csv['Distance_µm'] == -1) & (df_csv['Time'] == 1), 'Distance_µm'] = 0
            df_csv.loc[(df_csv['Speed_µm/min'] == -1) & (df_csv['Time'] == 1), 'Speed_µm/min'] = 0

            list_df.append(df_csv)
            list_fnames.append(f)

    # Affiche un message d'avertissement si DOUBLONS ou SAUT DE TEMPS
    check_issues(list_df, list_fnames, xp, scene, fiji=True)

    # On forme qu'un seul fichier pour une scène et on renomme les index
    df = pd.concat(list_df)
    df = df.sort_values(by = ['TrackID', 'Time'])
    df.reset_index(level = None, drop = True, inplace=True)

    # Ajout de la position relative
    df = relative_positions(df, unnamed = False, changeUnit = True)

    return df

def get_imaris_pos_dataframe(scene_dir: str, xp: str, scene: int) -> pd.DataFrame: 
    """
    Ouvre le csv IMARIS du paramètre 'Position' de scene_dir en vérifiant les problèmes
    de doublons et saut dans le temps, et en ajoutant les colonnes de positions relatives.
    
    Arguments:
        scene_dir (str): chemin vers le dossier de la scène contenant les csv
        xp (str): date de l'expérience
        scene (int): numéro de la scène

    Renvois: 
        pd.DataFrame: le dataframe décri ci-dessus
        """

    # Ouvre le fichier et calcule les positions relatives
    df_pos, filename = open_imaris_csv('Position', scene_dir)
    df_pos = relative_positions(df_pos) # a changer !

    # Affiche un message d'avertissement si DOUBLONS ou SAUT DE TEMPS
    check_issues([df_pos], filename, xp, scene) 

    # Ajoute les metadonnées
    df_pos = add_metadata_to_dataframe(df_pos, filename, df_meta=open_metadata())
    return df_pos

xp_dirs= ['P:\\Public\\Imagerie\\Videomicroscopie pièce 038\\sylvie\\2023\\2023-10-12']
scene_dir = xp_dirs[0]
scene = 6
res_regex = re.search(DATE_REGEX[0], scene_dir)
xp=res_regex.group(0) if res_regex else np.nan

df_pos = get_imaris_pos_dataframe(scene_dir=xp_dirs[0], xp=xp, scene=scene)
df_pos
```
